# Machine Learning Task Mapping

## Research Question
Which webpages should be refreshed first to maximize search performance?

## ML Task Type
Scoring

## Why This Task?
The goal is to assign each webpage an opportunity score based on historical search performance so that pages with the highest potential benefit can be refreshed first.

## Target
Opportunity Score

## Success Metric
Mean Absolute Error (MAE)
R² Score

## Key Concepts Learned
- AI vs Machine Learning vs Analytics vs Rule-Based Systems
- Machine Learning Loop
- Supervised vs Unsupervised Learning
- Generalization vs Overfitting
